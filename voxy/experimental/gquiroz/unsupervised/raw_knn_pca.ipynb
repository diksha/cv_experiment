# Feature Extraction


```python
def get_largest_common_img_size(image_paths):
    row_min = np.inf
    col_min = np.inf
    for image_path in image_paths:
        image = cv2.imread(image_path)
        row, col = image.shape[0:2]
        if row < row_min:
            row_min = row
        if col < col_min:
            col_min = col
    size = (row_min, col_min)
    return size
```

```python
# https://towardsdatascience.com/pca-using-python-scikit-learn-e653f8989e60
from sklearn.preprocessing import StandardScaler

import numpy as np
import cv2
import os

# Loading image paths
import glob
image_paths = glob.glob(DATASET_PATH + '/**/*.png', recursive=True)
image_paths.extend(glob.glob(DATASET_PATH + '/**/*.jpg', recursive=True))

DOOR_CLASSES = np.array(['open', 'partially_open', 'closed'])
STANDARD_IMG_SIZE = (32, 32)
HIGHEST_COMMON_SIZE = get_largest_common_img_size(image_paths)
RGB_LEN = 3
```

```python
def image_to_feature_vector(image, size=(32, 32)):
    # resize the image to a fixed size, then flatten the image into
    # a list of raw pixel intensities'
    return cv2.resize(image, (size[1], size[0])).flatten()
```

```python
def get_data_and_classes(images_path_list, img_size):
    X = np.zeros((len(images_path_list), img_size[0]*img_size[1]*RGB_LEN))
    y = np.zeros((len(images_path_list), 1)).flatten()
    for (i, image_path) in enumerate(images_path_list):
        image = cv2.imread(image_path)
        image_class = image_path.split(os.path.sep)[-2]
        class_id = np.where(DOOR_CLASSES == image_class)[0][0]
        pixels = image_to_feature_vector(image, img_size)
        X[i, :] = pixels
        y[i] = class_id
    return X, y
```

```python
def generate_scalar_transform(train_data):
    scaler = StandardScaler()
    return scaler.fit(train_data)
```

```python
def standardize_data(scalar_transform, X):
    # Apply transform to both the training set and the test set.
    return scalar_transform.transform(X)
```

```python
def extract_color_histogram(image, bins=(8, 8, 8)):
    # extract a 3D color histogram from the HSV color space using
    # the supplied number of `bins` per channel
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    hist = cv2.calcHist([hsv], [0, 1, 2], None, bins,
        [0, 180, 0, 256, 0, 256])
    
    # handle normalizing the histogram if we are using OpenCV 2.4.X
    hist = cv2.normalize(hist, hist)
    
    # return the flattened histogram as the feature vector
    return hist.flatten()
```

```python
# GET + STANDARDIZE DATA
train_images = [path for path in image_paths if 'train' in path]
test_images = [path for path in image_paths if 'test' in path]
ood_images = [path for path in image_paths if 'ood' in path]

train_pixels, train_class = get_data_and_classes(train_images, HIGHEST_COMMON_SIZE)
test_pixels, test_class = get_data_and_classes(test_images, HIGHEST_COMMON_SIZE)
ood_pixels, ood_class = get_data_and_classes(ood_images, HIGHEST_COMMON_SIZE)

scaler = generate_scalar_transform(train_pixels)
train_pixels_std = standardize_data(scaler, train_pixels)
test_pixels_std = standardize_data(scaler, test_pixels)
ood_pixels_std = standardize_data(scaler, ood_pixels)
```

# PCA

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
K_VAR_RETAINED = .99
```

```python
def plot_2_components(pca_ref, train_pca, classes):
    plt.figure(figsize=(10,8))
    closed_pca = train_pca[classes == np.where(DOOR_CLASSES == 'closed')[0][0]]
    po_pca = train_pca[classes == np.where(DOOR_CLASSES == 'partially_open')[0][0]]
    open_pca = train_pca[classes == np.where(DOOR_CLASSES == 'open')[0][0]]
    closed_scatter = plt.scatter(
        closed_pca[:, 0],
        closed_pca[:, 1],
        c='red',
    )
    po_scatter = plt.scatter(
        po_pca[:, 0],
        po_pca[:, 1],
        c='green',
    )
    open_scatter = plt.scatter(
        open_pca[:, 0],
        open_pca[:, 1],
        c='blue',
    )
    plt.legend([open_scatter, po_scatter, closed_scatter], DOOR_CLASSES)
    plt.title(f'N_COMPONENTS: {pca_ref.n_components_}')
    plt.xlabel('COMPONENT 1')
    plt.ylabel('COMPONENT 2')
    plt.show()
```

```python
def plot_image_reconstruction(original_image_mat, approx_image_mat, n_orig_features, n_components):
    plt.figure(figsize=(15,10))
    # Original Image
    plt.subplot(1, 2, 1);
    plt.imshow(
        original_image_mat,
    )
    plt.xlabel(f'{n_orig_features} components', fontsize = 14)
    plt.title('Original Image', fontsize = 20);

    # Reconstruction
    plt.subplot(1, 2, 2);
    plt.imshow(
        approx_image_mat,
    )
    plt.xlabel(f'{n_components} components', fontsize = 14)
    plt.title('PCA Reconstruction', fontsize = 20);
```

```python
pca = PCA(K_VAR_RETAINED)
pca.fit(train_pixels_std)
train_pixels_pca = pca.transform(train_pixels_std)
test_pixels_pca = pca.transform(test_pixels_std)
ood_pixels_pca = pca.transform(ood_pixels_std)
print(f"N COMPONENTS: {pca.n_components_}")
```

```python
# Reconstruction
train_pixels_approx_std = pca.inverse_transform(train_pixels_pca)
test_pixels_approx_std = pca.inverse_transform(test_pixels_pca)
ood_pixels_approx_std = pca.inverse_transform(ood_pixels_pca)
train_pixels_approx = scaler.inverse_transform(train_pixels_approx_std)
test_pixels_approx = scaler.inverse_transform(test_pixels_approx_std)
ood_pixels_approx = scaler.inverse_transform(ood_pixels_approx_std)
```

```python
orig_image = ood_pixels[20].reshape(HIGHEST_COMMON_SIZE[0], HIGHEST_COMMON_SIZE[1], RGB_LEN).astype(int)
n_orig_feat = ood_pixels.shape[1]
approx_image = ood_pixels_approx[20].reshape(HIGHEST_COMMON_SIZE[0], HIGHEST_COMMON_SIZE[1], RGB_LEN).astype(int)
n_components = pca.n_components_
plot_image_reconstruction(orig_image, approx_image, n_orig_feat, n_components)
```

```python
def plot_umap(umap_results, classes):
    plt.figure(figsize=(10,8))
    closed_umap = umap_results[classes == np.where(DOOR_CLASSES == 'closed')[0][0]]
    po_umap = umap_results[classes == np.where(DOOR_CLASSES == 'partially_open')[0][0]]
    open_umap = umap_results[classes == np.where(DOOR_CLASSES == 'open')[0][0]]
    closed_scatter = plt.scatter(
        closed_umap[:, 0],
        closed_umap[:, 1],
        c='red',
        
    )
    po_scatter = plt.scatter(
        po_umap[:, 0],
        po_umap[:, 1],
        c='green',
        
    )
    open_scatter = plt.scatter(
        open_umap[:, 0],
        open_umap[:, 1],
        c='blue',
    )
    plt.legend([open_scatter, po_scatter, closed_scatter], DOOR_CLASSES)
    plt.title(f'UMAP Embedding')
    plt.show()
```

```python
import random
import time

# KNN Parameters
NUM_NEIGHBORS = 5
NUM_JOBS = -1

# SVM Parameters
C = 1.0 # regularization
KERNEL = 'poly'
DEGREE = 3

# PCA PARAMS'
K_VAR_RETAINED = .99

# Model Dict
MODELS = {
    'KNN': KNeighborsClassifier(n_neighbors=NUM_NEIGHBORS, n_jobs=NUM_JOBS),
    'SVM': SVC(C=C, kernel=KERNEL, degree=DEGREE),
}
```

```python
MODEL = 'SVM'
dataset_percentage_sweep =  [0.1, 0.2, 0.5, 1.0]
acc_test = np.zeros(len(dataset_percentage_sweep))
acc_ood = np.zeros(len(dataset_percentage_sweep))
dataset_size = train_pixels_pca.shape[0]
for i, percentage in enumerate(dataset_percentage_sweep):
    # DATA PREPERATION
    k = int(dataset_size * percentage)
    print(f"----- Size of train set: {k}, from full set of size: {dataset_size}, Percentage: {percentage*100} -----")
    idx = np.random.choice(np.array(range(dataset_size)), size=k, replace=False)
    train_class_downsampled = train_class[idx]
    counts_dict = {DOOR_CLASSES[i]: np.sum(train_class_downsampled == i) for i in range(len(DOOR_CLASSES))}
    print(f"Count of each class: {counts_dict}")
    train_pixels_downsample = train_pixels[idx, :]
    scaler_transform = generate_scalar_transform(train_pixels_downsample)
    train_pixels_std_downsample = standardize_data(scaler_transform, train_pixels_downsample)
    pca = PCA(K_VAR_RETAINED)
    pca.fit(train_pixels_std_downsample)
    print(f"N Components: {pca.n_components_}")
    train_pixels_pca_downsample = pca.transform(train_pixels_std_downsample)
    
    # Model
    model_raw = MODELS[MODEL]
    model_raw.fit(train_pixels_pca_downsample, train_class_downsampled)
    
    # Inference TEST
    start_time = time.time()
    test_pixels_std = standardize_data(scaler_transform, test_pixels)
    test_pixels_pca = pca.transform(test_pixels_std)
    acc_test[i] = model_raw.score(test_pixels_pca, test_class)
    inference_time_test = time.time() - start_time 
    
    # Inference OOD
    start_time = time.time()
    ood_pixels_std = standardize_data(scaler_transform, ood_pixels)
    ood_pixels_pca = pca.transform(ood_pixels_std)
    acc_ood[i] = model_raw.score(ood_pixels_pca, ood_class)
    inference_time_ood = time.time() - start_time 
    print(f"Inference Time, Test: {inference_time_test}s, OOD: {inference_time_ood}s")
    
    # DRAW PLOTS
    y_pred = model_raw.predict(test_pixels_pca)
    print(f"Classes: {DOOR_CLASSES[model_raw.classes_.astype(int)]}")
    print(f"Per Class Precision {precision_score(test_class, y_pred, average=None)}")
    print(f"Per Class Recall {recall_score(test_class, y_pred, average=None)}")
    print(f"MISSING TEST LABELS: {set(test_class) - set(y_pred)}")
    cm = confusion_matrix(test_class, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=DOOR_CLASSES[model_raw.classes_.astype(int)])
    disp.plot()
    plt.title(f"{MODEL} With {percentage*100}% of Data")
    plt.show()
    
    # OOD
    y_pred = model_raw.predict(ood_pixels_pca)
    print(f"OOD Classes: {DOOR_CLASSES[model_raw.classes_.astype(int)]}")
    print(f"OOD: Per Class Precision {precision_score(ood_class, y_pred, average=None)}")
    print(f"OOD: Per Class Recall {recall_score(ood_class, y_pred, average=None)}")
    print(f"MISSING OOD LABELS: {set(ood_class) - set(y_pred)}\n\n")
```
